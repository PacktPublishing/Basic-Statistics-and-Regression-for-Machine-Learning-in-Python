```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#generate a normally distributed data set as an array
#generating 500 random values 
#arguments are mean,SD,no of samples required
heart_rates = np.random.normal(72,5,500)
print(heart_rates)
```

```python
# finding the z score of the heart rate array
z_scores = stats.zscore(heart_rates)
print(z_scores)
```

```python
# plot a historgram with heart rate over x axis and no of occurences over y axis
# creating a historgram with 60 bars in the plot
plt.hist(heart_rates,60)
plt.show()
```
